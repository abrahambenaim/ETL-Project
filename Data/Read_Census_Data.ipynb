```python
# SQLAlchemy
from sqlalchemy import create_engine

# Path to sqlite
database_path = "../Resources/Census_Data.sqlite"
```

```python
# Create an api engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
```

```python
# Query All Records in the the Database
header = engine.execute("SELECT * FROM Census_Data").keys()
print(header)
print("------------")
data = engine.execute("SELECT * FROM Census_Data WHERE state='NY'")
for record in data:
    print(record)
```
